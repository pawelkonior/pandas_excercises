```python
years = ['2016', '2017', '2018']

data = dt.loc[np.logical_and(
    np.logical_and(
        dt['product_types'] == 'sok jabłkowy, w opakowaniu kartonowym - za 1l',
        dt['province'] == 'MAZOWIECKIE'
    ),
    dt['date'].str[:4].isin(years)
)]

data['value'].mean()
```

```python
years = [str(year) for year in range(2003, 2016)]

data = dt.loc[np.logical_and(
    np.logical_and(
        dt['product_types'] == 'koncentrat pomidorowy 30% - za 1kg',
        dt['province'] == 'DOLNOŚLĄSKIE'
    ),
    np.logical_and(
        dt['date'].str[:4].isin(years),
        dt['value'] > 0
    )
)]

data.loc[data['value'] != 3000].value.mean()
# data['value'].mean()
```
