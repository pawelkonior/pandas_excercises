```python
df = pd.read_csv(os.path.join('..', 'datasets', 'ceny_produktow_cleaned.csv'), sep=";")
df.head()
```

```python
df = df.loc[df['province'] != 'POLSKA']
df.sample(5)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df['year'] = df['date'].dt.year
```

```python
df_grouped = df.groupby(by=['product', 'year'])
```

```python
df_grouped[['value']].agg(['min', 'max', 'median', 'std'])
```

```python
for key, group_df in df_grouped:
    print(key, group_df[['value']].agg('min'))
```
