```python
df_by_id = df.groupby(by=['product_group_id'])
```

```python
len(df_by_id)
```

```python
wb = openpyxl.Workbook()

for idx, group in df_by_id:
    ws = wb.create_sheet(str(idx[0]), idx[0] - 1)
    
    for row in dataframe_to_rows(group, index=False, header=True):
        ws.append(row)

wb.save(os.path.join(ds_path, 'ceny_produktow_groupped.xlsx'))
```

```python
ds_path = os.path.join('..', 'datasets')

df = pd.read_csv(
    os.path.join(ds_path, 'ceny_produktow_cleaned.csv'),
    sep=';',
    parse_dates=['date']
)
```

```python
df['date_formatted'] = df['date'].dt.strftime('%Y-%m')
```

```python
group_filter = df['product_group_id'] == (group_id if group_id is not None else df['product_group_id'])
product_filter = df['product'] == (product if product is not None else df['product'])
date_filter = df['date'] == (date_ if date_ is not None else df['date'])
value_filter = df['value'] > 0
```

```python
df_report = df.loc[group_filter & product_filter & date_filter & value_filter]
```

```python
df_report.head()
```

```python
pt = pd.pivot_table(
    data=df_report,
    index=['province', 'product'],
    columns=['date_formatted'],
    values='value',
    aggfunc='mean'
)
```

```python
pt.head()
```

```python
pt.to_excel(os.path.join(ds_path, 'exercise_3.xlsx'))
```
