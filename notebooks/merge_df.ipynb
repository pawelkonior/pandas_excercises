```python
df_joined.loc[:50, ["value_x", "value_y", "date_x", "date_y"]]
```

```python
import numpy as np
```

```python
df_joined = df_joined.loc[np.logical_and(
    df_joined['value_x'] >= df_joined['value_y'],
    df_joined['date_x'] > df_joined['date_y']
)]
```
