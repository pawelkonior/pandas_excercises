```python
# is_palindrome (kajak, Ala)
def is_palindrome(text):
    return text.lower() == text.lower()[::-1]
```

```python
is_palindrome2 = lambda text: text.lower() == text.lower()[::-1]
```

```python
is_palindrome2('()()')
```

```python
def magic(x):
    res = sorted(x, reverse=True)[2] 
    if pd.isna(res):
        return 0
    return res

pd.pivot_table(
    data=df,
    index=['product'],
    columns=['province'],
    values=['value'],
    aggfunc=magic
)
```

```python
pd.pivot_table(
    data=df,
    index=['product'],
    columns=['year'],
    values=['value'],
    aggfunc=['mean', 'median']
)
```
